### Hyrje
Ky raport analizon se si ndryshon performanca e kompjuterit kur rritet madhësia e punës që i kërkohet sistemit. Performanca matet përmes kohës së ekzekutimit të një detyre llogaritëse. Në praktikë, kjo kohë mund të ndikohet nga procesori (CPU), memoria RAM, proceset e tjera aktive, sistemi operativ dhe temperatura e kompjuterit.

Qëllimi i eksperimentit është të krijohet një matje e thjeshtë në Python dhe të paraqitet variacioni i performancës me tabelë dhe grafik.


```python
import time
import platform
import statistics
import numpy as np
import matplotlib.pyplot as plt

print("Informacion bazë për sistemin:")
print("Sistemi Operativ:", platform.system(), platform.release())
print("Procesori:", platform.processor() or "Nuk u identifikua nga Python")
print("Versioni i Python:", platform.python_version())
```

## 1. Përshkrimi i eksperimentit

Në këtë eksperiment përdoret një funksion që kryen shumë veprime aritmetike. Sa më e madhe të jetë vlera `n`, aq më shumë punë duhet të kryejë kompjuteri. Për çdo vlerë të `n`, matja përsëritet disa herë dhe pastaj llogariten:

- koha mesatare e ekzekutimit;
- devijimi standard, që tregon sa ndryshojnë matjet nga njëra-tjetra;
- variacioni i performancës nga një madhësi pune te tjetra.


```python
def detyre_llogaritese_donaldo(n):
    """Funksion testues për ngarkesë CPU.
    Kryen operacione numerike për të matur kohën e ekzekutimit.
    """
    total = 0
    for i in range(n):
        total += (i * i + 3 * i) % 97
    return total

# Vlera të ndryshme për madhësinë e punës
n_vlerat = [10_000, 30_000, 60_000, 100_000, 200_000, 400_000, 800_000]

kohet_mesatare = []
devijimet_standarde = []
prova_per_cdo_n = 5

for n in n_vlerat:
    prova = []
    for _ in range(prova_per_cdo_n):
        start = time.perf_counter()
        detyre_llogaritese_donaldo(n)
        end = time.perf_counter()
        prova.append(end - start)
    
    kohet_mesatare.append(statistics.mean(prova))
    devijimet_standarde.append(statistics.stdev(prova))

print("Matjet u kryen me sukses.")
```

## 2. Paraqitja e të dhënave

Tabela më poshtë tregon kohën mesatare të ekzekutimit për secilën madhësi pune. Nëse koha rritet kur rritet `n`, atëherë kompjuteri po harxhon më shumë burime për të përfunduar detyrën.


```python
print(f"{'n':>12} | {'Koha mesatare (s)':>20} | {'Devijimi standard':>20}")
print("-" * 60)
for n, mes, dev in zip(n_vlerat, kohet_mesatare, devijimet_standarde):
    print(f"{n:>12} | {mes:>20.6f} | {dev:>20.6f}")
```

## 3. Grafiku i variacionit të performancës

Grafiku tregon lidhjen midis madhësisë së punës dhe kohës mesatare të ekzekutimit. Sa më e pjerrët të jetë kurba, aq më shumë rritet koha kur rritet ngarkesa.


```python
plt.figure(figsize=(8, 5))
plt.plot(n_vlerat, kohet_mesatare, marker='o')
plt.title("Variacioni i Performancës së Kompjuterit")
plt.xlabel("Madhësia e punës (n)")
plt.ylabel("Koha mesatare e ekzekutimit (sekonda)")
plt.grid(True)
plt.show()
```

## 4. Analizë log-log

Për të kuptuar më mirë sjelljen e performancës, përdoret edhe një grafik në shkallë logaritmike. Nëse grafiku është afërsisht vijë e drejtë, atëherë rritja e kohës ndjek një model të rregullt matematikor.


```python
log_n = np.log(n_vlerat)
log_koha = np.log(kohet_mesatare)

koeficientet = np.polyfit(log_n, log_koha, 1)
pjerrësia = koeficientet[0]

plt.figure(figsize=(8, 5))
plt.plot(log_n, log_koha, marker='o')
plt.title("Analiza log-log e performancës")
plt.xlabel("log(n)")
plt.ylabel("log(koha)")
plt.grid(True)
plt.show()

print("Pjerrësia e përafërt:", round(pjerrësia, 3))
```

## 5. Interpretimi i rezultateve

Nga matjet vërehet se koha e ekzekutimit rritet kur rritet madhësia e punës `n`. Kjo tregon se performanca e kompjuterit nuk është konstante, por varet nga ngarkesa që i jepet sistemit.

Variacionet e vogla midis provave mund të shkaktohen nga:

- programe të tjera që punojnë në sfond;
- përdorimi momental i CPU-së;
- menaxhimi i memories nga sistemi operativ;
