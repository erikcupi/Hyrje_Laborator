## Qëllimi i Laboratorit
Ky laborator synon të demonstrojë hapat kryesorë të menaxhimit të versioneve duke përdorur **Git** lokal dhe platformën **GitHub** për bashkëpunim.

### Çfarë është Git dhe GitHub?
* **Git**: Një sistem i shpërndarë për kontrollin e versioneve (VCS) që gjurmon ndryshimet në kod.
* **GitHub**: Një platformë në cloud që lejon ruajtjen e repositorëve Git dhe bashkëpunimin me të tjerët.

### Inicializimi i një Repositori të ri
Për të filluar ndjekjen e një projekti, përdorim komandën `git init`. Kjo krijon një direktori të fshehur `.git`.

## Inicializojmë projektin lokal
!git init

## Kontrollojmë gjendjen e skedarëve
!git status

### Shtimi i ndryshimeve (Add & Commit)
Procesi i punës ndjek këtë rrugëtim:
1. **Working Directory**: Skedarët ku po punoni.
2. **Staging Area**: Skedarët që bëhen gati për t'u ruajtur (`git add`).
3. **Local Repository**: Ndryshimet ruhen në historik (`git commit`).

## Kriojmë një skedar të ri për test
!echo "Kodi i parë i projektit" > programi.py

## 1. Kalojmë skedarin në Staging Area
!git add programi.py

## 2. Kryejmë Commit-in
!git commit -m "Laboratori 02: Shtimi i skedarit programi.py"

## Shohim historikun e veprimeve
!git log --oneline

### Publikimi në GitHub
Për të dërguar punën tonë në cloud, duhet të lidhim projektin lokal me një "Remote Repository" në GitHub.

## Përfundime dhe Reflektime (LAB 02)

Përmes këtij laboratori, kemi dokumentuar dhe praktikuar ciklin jetësor të një projekti duke përdorur **Git** dhe **GitHub**. Pikat kryesore të trajtuara janë:

### 1. Menaxhimi i Versioneve (Version Control)
Kemi parë se si Git na lejon të mbajmë një "snapshot" të çdo ndryshimi. Kjo është jetike për:
* **Sigurinë e kodit**: Mundësia për t'u kthyer në një version të mëparshëm nëse diçka shkon keq.
* **Historikun**: Dokumentimi i asaj që është bërë përmes mesazheve të `commit`.

### 2. Diferenca midis Git dhe GitHub
Gjatë punës u bë e qartë se:
* **Git** është mjeti lokal (motori) që bën gjurmimin e skedarëve.
* **GitHub** është platforma në re (cloud) që shërben si urë bashkëpunimi dhe backup-i.

### 3. Workflow-i Standard
Procesi i ndjekur nga terminali i Jupyter-it (`add` -> `commit` -> `push`) është standardi i industrisë që çdo zhvillues softueri përdor çdo ditë.
