## Qëllimi
Të llogaritet dendësia e një objekti dhe pasiguria e saj.

```python
import numpy as np

# Masa (g)
m = np.array([78.42, 78.39, 78.44, 78.41, 78.43])

# Përmasat (cm)
a = np.array([5.02, 5.01, 5.03])
b = np.array([2.51, 2.50, 2.52])
c = np.array([1.25, 1.26, 1.25])

# Mesataret
m_mes = np.mean(m)
a_mes = np.mean(a)
b_mes = np.mean(b)
c_mes = np.mean(c)

print("Mesataret:")
print("m =", m_mes)
print("a =", a_mes)
print("b =", b_mes)
print("c =", c_mes)
```

```python
# Pasiguritë
dm = 0.01
da = 0.01
db = 0.01
dc = 0.01

# Vëllimi
V = a_mes * b_mes * c_mes

# Pasiguria e V
rel_V = da/a_mes + db/b_mes + dc/c_mes
dV = V * rel_V

print("V =", V, "+/-", dV)
```

```python
# Dendësia
rho = m_mes / V

# Pasiguria e dendësisë
rel_rho = dm/m_mes + rel_V
drho = rho * rel_rho

print("Dendësia:")
print("rho =", rho, "+/-", drho, "g/cm^3")
```
