## 1. Qëllimi i Punës Laboratorike

Qëllimi i këtij laboratori është njohja praktike me instrumentet kryesore që përdoren për matjen e gjatësisë dhe masës. 
Përmes kësaj detyre, synojmë:
* Të kuptojmë parimin e punës së secilit instrument.
* Të përcaktojmë vlerën më të vogël të shkallës (VMS) për secilin mjet.
* Të kryejmë matje direkte mbi objekte të ndryshme dhe të dokumentojmë saktësinë e tyre.

## 2. Instrumentet e përdorura dhe funksioni i tyre

Gjatë këtij laboratori kemi përdorur pajisjet e mëposhtme:

1. **Vizorja Milimetrike:**
   * *Përdorimi:* Shërben për matjen e gjatësive të thjeshta dhe të drejta ku nuk kërkohet saktësi e lartë.
   * *Saktësia:* Zakonisht $1 \text{ mm}$.

2. **Kalibri:**
   * *Përdorimi:* Instrument i gjithanshëm që shërben për matjen e diametrave të jashtëm, diametrave të brendshëm (p.sh. gryka e një epruvete) dhe thellësisë.
   * *Saktësia:* $0.1 \text{ mm}$ ose $0.05 \text{ mm}$.

3. **Mikrometri:**
   * *Përdorimi:* Shërben për matjen e dimensioneve shumë të vogla si trashësia e një flete, diametri i një teli ose i një sferë të vogël metalike.
   * *Saktësia:* Shumë e lartë, deri në $0.01 \text{ mm}$.

4. **Sferometri:**
   * *Përdorimi:* Përdoret specifikisht për të matur rrezen e kurbaturës së sipërfaqeve sferike ose trashësinë e objekteve të rrafshëta e të holla.

5. **Peshorja Elektronike:**
   * *Përdorimi:* Shërben për përcaktimin e masës së objekteve me një lexim të shpejtë digjital.
   * *Saktësia:* Varion nga $0.1 \text{ g}$ deri në $0.01 \text{ g}$ (sipas modelit).

6. **Peshorja Manuale:**
   * *Përdorimi:* Përcaktimi i masës përmes krahasimit të peshave etalon në dy krahët e peshores deri në arritjen e ekuilibrit.

### D. Matja e Masës (Peshore Elektronike)

### F. Matja me Mikrometër

## 5. Konkluzione dhe Diskutim

Pas përfundimit të matjeve në këtë laborator, arritëm në këto konkluzione kryesore:

* *Saktësia e Instrumenteve:* Konfirmuam eksperimentalisht se saktësia e matjes rritet kur përdorim instrumente me VMS (vlerë më të vogël të shkallës) më të vogël. Ndërsa vizorja na lejon matje deri në $1 \text{ mm}$, kalibri dhe mikrometri na mundësuan leximin e fraksioneve të milimetrit me saktësi deri në $0.01 \text{ mm}$.
* *Zbatueshmëria:* Vumë re se për objekte me gjeometri komplekse, si globi dekorativ apo lentet, përdorimi i sferometrit është i domosdoshëm për të përcaktuar kurbaturën e sipërfaqes, matje kjo që nuk mund të bëhet me instrumente të thjeshta lineare.
* *Analiza e Masës:* Krahasimi midis peshores manuale dhe asaj elektronike tregoi se, ndonëse të dyja metodat janë të sakta, peshorja elektronike minimizon gabimin njerëzor gjatë leximit dhe kursen kohë gjatë procesit të peshimit të objekteve të vogla si kufjet.
* *Gabimet në Matje:* Identifikuam se pasiguria në matje nuk varet vetëm nga instrumenti, por edhe nga faktorë si presioni i tepërt (te mikrometri) apo pozicioni jo-perpendikular i vështrimit (gabimi i paralaksit), të cilat u përpoqëm t'i minimizojmë gjatë këtij laboratori.
