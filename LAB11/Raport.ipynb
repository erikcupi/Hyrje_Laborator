## Qëllimi

Qëllimi i këtij raporti është të studiohet saktësia e GPS/GNSS dhe variacioni i pozicionit gjatë matjeve. Edhe kur pajisja qëndron në të njëjtin vend, koordinatat mund të ndryshojnë pak nga njëra matje në tjetrën.

## Hyrje

GPS është një sistem navigimi satelitor që përdoret për të përcaktuar pozicionin në Tokë. GNSS është term më i gjerë që përfshin GPS dhe sisteme të tjera satelitore.

Saktësia e GPS/GNSS mund të ndikohet nga:

- pengesat fizike si ndërtesa dhe pemë;
- kushtet atmosferike;
- reflektimi i sinjaleve;
- numri i satelitëve;
- cilësia e pajisjes matëse.

```python
# Te dhena te simuluara GPS/GNSS
# Cdo element perfaqeson nje matje ne te njejtin vend

latitude = [
    41.32751, 41.32748, 41.32752, 41.32750, 41.32749,
    41.32753, 41.32747, 41.32750, 41.32751, 41.32749
]

longitude = [
    19.81871, 19.81868, 19.81872, 19.81870, 19.81869,
    19.81873, 19.81867, 19.81870, 19.81871, 19.81869
]

print("Matjet u ruajten me sukses.")
print("Numri i matjeve:", len(latitude))
```

```python
# Shfaqja e te dhenave

print("Nr | Latitude  | Longitude")
print("---------------------------")

for i in range(len(latitude)):
    print(i + 1, "|", latitude[i], "|", longitude[i])
```

## Llogaritja e Pozicionit Mesatar

Pozicioni mesatar gjendet duke llogaritur mesataren e të gjitha vlerave të latitude dhe longitude. Ky pozicion konsiderohet si vlerësimi më i mirë i vendndodhjes.

```python
mean_lat = sum(latitude) / len(latitude)
mean_lon = sum(longitude) / len(longitude)

print("Latitude mesatare:", mean_lat)
print("Longitude mesatare:", mean_lon)
```

## Devijimi nga Pozicioni Mesatar

Për çdo matje llogaritet sa larg është ajo nga pozicioni mesatar. Kjo ndihmon për të kuptuar variacionin e pozicionit.

```python
deviation_lat = []
deviation_lon = []

for i in range(len(latitude)):
    deviation_lat.append(latitude[i] - mean_lat)
    deviation_lon.append(longitude[i] - mean_lon)

print("Nr | Devijimi Latitude | Devijimi Longitude")
print("---------------------------------------------")

for i in range(len(latitude)):
    print(i + 1, "|", round(deviation_lat[i], 8), "|", round(deviation_lon[i], 8))
```

## Llogaritja e Devijimit Standard

Devijimi standard tregon sa të shpërndara janë matjet rreth mesatares. Sa më i vogël devijimi standard, aq më e saktë është matja.

```python
def standard_deviation(values):
    mean_value = sum(values) / len(values)
    squared_differences = []

    for value in values:
        squared_differences.append((value - mean_value) ** 2)

    variance = sum(squared_differences) / len(values)
    return variance ** 0.5

std_lat = standard_deviation(latitude)
std_lon = standard_deviation(longitude)

print("Devijimi standard i Latitude:", std_lat)
print("Devijimi standard i Longitude:", std_lon)
```

## Vizualizimi i Thjeshtë i Matjeve

Më poshtë shfaqet një grafik i thjeshtë tekstual, ku çdo matje krahasohet me pozicionin mesatar.

```python
print("Grafik tekstual i ndryshimit te Latitude")
print("----------------------------------------")

for i in range(len(latitude)):
    diff = abs(latitude[i] - mean_lat)
    stars = int(diff * 1000000)
    print("Matja", i + 1, ":", "*" * stars)
```

Pozicioni mesatar është më i besueshëm sesa një matje e vetme, sepse ai përfaqëson qendrën e të gjitha matjeve të marra.

Devijimi standard është i vogël, prandaj matjet mund të konsiderohen relativisht të sakta.
