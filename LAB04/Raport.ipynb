# LAB04 - Python Basics

This notebook covers:
- Data types
- String manipulation (f-strings)
- Functions
- Flow control (if-else, for loops)
- NumPy basics
- Matplotlib visualization

## 1. Data Types

```
# Basic data types
integer_num = 10
float_num = 3.14
string_text = "Hello, Python!"
boolean_val = True

print(type(integer_num))
print(type(float_num))
print(type(string_text))
print(type(boolean_val))
```

## 2. String Manipulation (f-strings)

```
name = "Donaldo"
age = 18

message = f"My name is {name} and I am {age} years old."
print(message)
```

## 3. Functions

```
def square_number(x):
    return x ** 2

print(square_number(5))
```

## 4. Flow Control (if-else, for loop)

```
# if-else example
num = 7

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")

# for loop example
for i in range(5):
    print(f"Iteration {i}")
```

## 5. NumPy Basics

```
import numpy as np

array = np.array([1, 2, 3, 4, 5])
print("Array:", array)
print("Mean:", np.mean(array))
print("Sum:", np.sum(array))
```

## 6. Matplotlib Visualization

```
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = x ** 2

plt.plot(x, y)
plt.title("Simple Plot: y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
