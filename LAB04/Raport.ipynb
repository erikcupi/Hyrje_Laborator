## 2. f-strings Example

## 3. Simple Function

```
def add_numbers(x, y):
    return x + y

result = add_numbers(3, 4)
print("Sum:", result)
```

## 4. Condition and Loop

```
numbers = [1, 2, 3, 4, 5]

for n in numbers:
    if n > 3:
        print(f"{n} is greater than 3")
    else:
        print(f"{n} is 3 or less")
```

## 5. NumPy Example

```
import numpy as np

arr = np.array([10, 20, 30])
print("Max:", np.max(arr))
print("Min:", np.min(arr))
```

## 6. Simple Plot

```
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
plt.title("Linear Graph")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```
